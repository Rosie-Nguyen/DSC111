```python
N = len(df2)

def getLocalMaximum(series, seuil):
    series_array = series.values
    # Lay trung binh dieu hoa
    harmonized_array = (series_array + np.append(series_array[1:], 0) + np.append(0, series_array[: N-1]))/3
    localMax = []
    for i, s in enumerate(series[1 : N-1]):
        if harmonized_array[i] - harmonized_array[i-1] >= seuil and harmonized_array[i] - harmonized_array[i+1] >= seuil:
            localMax.append(i)
    return np.array(localMax)

localMaximum = getLocalMaximum(prices['SE1'], 0.1)
localMaximum
```

## Linear Regression

```python
DEPENDENCY_PERIOD = 50

def makeDataFile(priceDictionary = df2, code='SE1', dependencyPeriod = DEPENDENCY_PERIOD, futurePredictionTime = 0, mode = 'price'):
    series = priceDictionary[code].values
    dataLength = len(series) - dependencyPeriod - futurePredictionTime
    X = np.array([[series[i + j] for j in range(dependencyPeriod)] for i in range(dataLength)])
    if (mode == 'variation'):
        y = np.array(series[dependencyPeriod + futurePredictionTime:] - series[dependencyPeriod - 1: -1]) 
    else:
        y = np.array(series[dependencyPeriod + futurePredictionTime:])
    beginning_time = datetime[:dataLength].values
    end_time = datetime[dependencyPeriod - 1: dependencyPeriod + dataLength - 1].values
    prediction_time = datetime[dependencyPeriod + futurePredictionTime:].values
    assert(len(X) == len(y))
    return X, y, beginning_time, end_time, prediction_time 

X, y, t1, t2, t3 = makeDataFile(prices, 'SE1', 50, 0, mode='price')
pd.DataFrame(X).head()
```

```python
pd.DataFrame(y).head()
```

```python
from sklearn.cross_validation import cross_val_score
TEST_SPLIT = 0.2
VALIDATION_SPLIT = 0.2 

np.random.seed(0)
splitter = int((1 - TEST_SPLIT) * len(X))
X_train, X_test = X[:splitter], X[splitter:]
y_train, y_test = y[:splitter], y[splitter:]
t1_train, t2_train, t3_train, t1_test, t2_test, t3_test = t1[:splitter], t2[:splitter], t3[:splitter], t1[splitter:], t2[splitter:], t3[splitter:]
```

```python
def makeDataFile2(priceDictionary = df2, code='SE1', dependencyPeriod = DEPENDENCY_PERIOD, futurePredictionTime = 0, mode = 'price'):
    series = priceDictionary[code].values
    dataLength = len(series) - dependencyPeriod - futurePredictionTime
    Z = np.array([[series[i + j] for j in range(dependencyPeriod)] for i in range(dataLength)])
    X = np.concatenate([hours.values[:dataLength], Z], axis = 1)
    if (mode == 'variation'):
        y = np.array(series[dependencyPeriod + futurePredictionTime:] - series[dependencyPeriod - 1: -1]) 
    else:
        y = np.array(series[dependencyPeriod + futurePredictionTime:])
    beginning_time = datetime[:dataLength].values
    end_time = datetime[dependencyPeriod - 1: dependencyPeriod + dataLength - 1].values
    prediction_time = datetime[dependencyPeriod + futurePredictionTime:].values
    assert(len(X) == len(y))
    return X, y, beginning_time, end_time, prediction_time
```
